```python
import pandas as pd
```

```python
superstore=pd.read_csv("../Superstore.csv", sep=",", encoding="latin-1")
superstore
```

```python
Anzahl=superstore["Category"].count()
print(Anzahl)#9994 Zeilen
kategorien=superstore["Category"].value_counts()
rel_kategorien=(kategorien[["Furniture","Office Supplies","Technology"]]/9994*100).round(1)

rel_kategorien.plot(kind="pie",ylabel="")

```
