```python
import pandas as pd
```

```python
superstore=pd.read_csv(r"C:\Users\jan\Documents\PythonKurs\07_Python_für_Datenanalysten\Daten_Analyse_Projekt\Superstore.csv", sep=",", encoding="latin-1")
superstore.head(5)
```

```python
Anzahl=superstore["Category"].count()
kategorien=superstore["Category"].value_counts()
rel_kategorien=(kategorien[["Furniture","Office Supplies","Technology"]]/Anzahl*100).round(1)
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0,0.2),title=f"Main Categories (Absolute data: {Anzahl} shipments)")
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0.2,0,0),title=f"Main Categories (Absolute data: {Anzahl} shipments)")
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0.2,0),title=f"Main Categories (Absolute data: {Anzahl} shipments)",startangle=180)
```
