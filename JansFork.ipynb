```python
import pandas as pd
from matplotlib import colormaps
```

```python
superstore=pd.read_csv("superstore_cleaned.csv", sep=",", encoding="latin-1")
superstore.head(5)
```

```python
Anzahl=superstore["Category"].count()
kategorien=superstore["Category"].value_counts()
rel_kategorien=(kategorien[["Furniture","Office Supplies","Technology"]]/Anzahl*100).round(1)
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0,0.2),title=f"Main Categories (Absolute data: {Anzahl} shipments)")
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0.2,0,0),title=f"Main Categories (Absolute data: {Anzahl} shipments)")
```

```python
rel_kategorien.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0.2,0),title=f"Main Categories (Absolute data: {Anzahl} shipments)",startangle=180)
```

```python
#print(superstore.groupby("Category")["Sub-Category"].nunique()) #--> 17 Kategorien
unterkategorie=superstore['Sub-Category'].unique()

Kategorie1=superstore["Category"]=="Technology"
Kategorie2=superstore["Category"]=="Furniture"
Kategorie3=superstore["Category"]=="Office Supplies"

AnzahlTech=superstore[Kategorie1]["Sub-Category"].count()
EintraegeTech=superstore[Kategorie1]["Sub-Category"].nunique()
Technology=superstore[Kategorie1]["Sub-Category"].value_counts()
Technology.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0,0,0),title=f"Sub-Category: Technology (Absolute data: {AnzahlTech} shipments)",startangle=0,colormap="Greens_r")#,colors = pd.Pastel1(range(len(Technology))))

```

```python
AnzahlFurniture=superstore[Kategorie2]["Sub-Category"].count()
EintraegeFurniture=superstore[Kategorie2]["Sub-Category"].nunique()
Furniture=superstore[Kategorie2]["Sub-Category"].value_counts()
Furniture.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0,0,0),title=f"Sub-Category: Furniture (Absolute data: {AnzahlFurniture} shipments)",startangle=0,colormap="Blues_r")
```

```python
AnzahlOffice=superstore[Kategorie3]["Sub-Category"].count()
EintraegeOffice=superstore[Kategorie3]["Sub-Category"].nunique()
OfficeSupplies=superstore[Kategorie3]["Sub-Category"].value_counts()
OfficeSupplies.plot(kind="pie",ylabel="",autopct="%1.1f%%",explode=(0,0,0,0,0,0,0,0,0),title=f"Sub-Category: Office Supplies (Absolute data: {AnzahlOffice} shipments)",startangle=0,colormap="Oranges_r")
print(len(superstore))
print(AnzahlTech+AnzahlOffice+AnzahlFurniture)
print("PrÃ¼fung i.O. Anzahlen der GegenstÃ¤nde in Kategorien sind gleich den Anzahlen der Zeilen")
#list(colormaps)
```
