```python
superstore["Product Name"].value_counts() #am meisten werden "Staples" gekauft
TopdreiVerkauft=superstore["Product Name"].value_counts().iloc[:3].sort_values(ascending=True)
print(TopdreiVerkauft)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(TopdreiVerkauft.index,TopdreiVerkauft)#width=0.6)

# Set individual colors for each bar
colors = ['#1f77b4', "#ff9900", "#068706"]  # Dark blue, Orange, Green
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Customize the plot
ax.set_title('Top three most selled products', pad=20,fontweight="bold")
ax.set_xlabel('Quantity selled', labelpad=20)
ax.set_ylabel("", labelpad=20)

#wert an den jeweiligen balken schreiben:
for i, x in enumerate(TopdreiVerkauft):
    ax.text(x+1, i+0, f'{x:,.0f} ', va='center', fontsize=10)
#TopdreiVerkauft.plot(kind="bar",title=f"Top ten most selled products",rot=90,ylabel="Quantity Selled")

```

```python

#PosProfit=superstore["Profit"]>0
#NegProfit=superstore["Profit"]<=0
Pos_Kategorie=superstore.groupby("Sub-Category")["Profit"].sum().sort_values().round(2)
Pos_Kategorie=Pos_Kategorie/1000
Pos_Kategorie.round(2)
Neg_Kategorie=superstore.groupby("Sub-Category")["Profit"].sum().sort_values().round(2)
Neg_Kategorie=Neg_Kategorie*(-1)/1000

#print(Pos_Kategorie)
#Pos_Kategorie.plot(kind="line",x="Product Name",y="Profit")

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(Pos_Kategorie.index,Pos_Kategorie)#, width=0.6)

# Set individual colors for each bar
colors = ["red","red","red"]#'#1f77b4', "#ff9900", "#068706"]  # Dark blue, Orange, Green
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Customize the plot
ax.set_title('Profit of "Sub-Categories"', pad=20,fontweight="bold")
ax.set_xlabel('Profit in 1000 $', labelpad=20,fontweight="bold")
ax.set_ylabel("", labelpad=20,fontweight="bold")

#wert an den jeweiligen balken schreiben:
for i, x in enumerate(Pos_Kategorie):
    ax.text(x+0.1, i+0, f'{x:,.1f} ', va='center', fontsize=10)
```
